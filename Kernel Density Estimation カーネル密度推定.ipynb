```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
lambda_A = 5
lambda_B = 15

num_A = 250
num_B = 500
num = num_A + num_B

A = np.random.poisson(lambda_A, num_A)
B = np.random.poisson(lambda_B, num_B)

X = np.concatenate([A, B])

plt.figure(figsize=(12, 5))
plt.title('Multiple Poisson Distributions')
plt.hist(X, bins = 25)
plt.show()
```

```python
sns.kdeplot(X)
```

```python
Xnum = 100
h = 1

x = np.linspace(-5, 40, Xnum)

px = np.zeros([1, Xnum])

for i in X:
    fx = (2 * np.pi)**(-0.5) * np.exp(-0.5 * (x - i)**2 / h**2)
    px = px + fx
    
px = px / (num * h)

plt.plot(x, px[0])
```

```python
sns.kdeplot(X)
plt.plot(x, px[0])
```
